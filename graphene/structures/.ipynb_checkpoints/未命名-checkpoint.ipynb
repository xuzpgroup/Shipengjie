```python
from ase.io import read, write
import numpy as np
import matplotlib.pyplot as plt
```

```python
atom = read('9_0_edge.vasp')
pos = atom.get_positions()
cell = np.array(atom.get_cell())
```

```python
plt.figure(figsize=(10,10))
com = [0,0]
Th_norm = cell[0,0]
Tv_norm = cell[1,1]
plt.plot([com[0], Th_norm + com[0]], [com[1], com[1]], 'r')
plt.plot([com[0], com[0]], [com[1], com[1] + Tv_norm], 'r')
plt.plot([com[0], Th_norm + com[0]], [com[1] + Tv_norm, com[1] + Tv_norm], 'r')
plt.plot([com[0] + Th_norm, com[0] + Th_norm], [com[1], com[1] + Tv_norm], 'r')
plt.scatter(pos[:,0],pos[:,1])


"""
min_row, min_col, max_row, max_col = 38.5,33.5,40.5,35.5
cat_dict = (min_row, min_col, max_row, max_col)
plt.plot([cat_dict[1], cat_dict[3], cat_dict[3], cat_dict[1], cat_dict[1]], # col
     [cat_dict[0], cat_dict[0], cat_dict[2], cat_dict[2], cat_dict[0]], # row
     color='b')
I = np.argwhere(((pos[:,0]>min_col) & (pos[:,0]<max_col)) &
                  ((pos[:,1]>min_row) & (pos[:,1]<max_row)))[0,0]
print(I)
"""
plt.axis('square')
plt.axis([-5,Th_norm+5,-5,Tv_norm+5])
```

```python
pos[I][0] -= cell[0,0]
```

```python
plt.figure(figsize=(4,8))
com = [0,0]
Th_norm = cell[0,0]
Tv_norm = cell[1,1]
plt.plot([com[0], Th_norm + com[0]], [com[1], com[1]], 'r')
plt.plot([com[0], com[0]], [com[1], com[1] + Tv_norm], 'r')
plt.plot([com[0], Th_norm + com[0]], [com[1] + Tv_norm, com[1] + Tv_norm], 'r')
plt.plot([com[0] + Th_norm, com[0] + Th_norm], [com[1], com[1] + Tv_norm], 'r')
plt.scatter(pos[:,0],pos[:,1])
plt.scatter(pos[:,0],pos[:,1]+Tv_norm)
plt.axis('square')
plt.axis([-5,Th_norm+5,-5,Tv_norm+15])
```

```python
new_pos = pos
new_pos[:,0] += 50 - cell[0,0]/2
new_cell = cell
new_cell[0,0]=100
new_atom = atom.copy()
new_atom.set_positions(new_pos)
new_atom.set_cell(new_cell)
```

```python
plt.figure(figsize=(8,8))
com = [0,0]
Th_norm = new_cell[0,0]
Tv_norm = new_cell[1,1]
plt.plot([com[0], Th_norm + com[0]], [com[1], com[1]], 'r')
plt.plot([com[0], com[0]], [com[1], com[1] + Tv_norm], 'r')
plt.plot([com[0], Th_norm + com[0]], [com[1] + Tv_norm, com[1] + Tv_norm], 'r')
plt.plot([com[0] + Th_norm, com[0] + Th_norm], [com[1], com[1] + Tv_norm], 'r')
plt.scatter(new_pos[:,0],new_pos[:,1])
plt.axis('square')
plt.axis([-5,Th_norm+5,-5,Tv_norm+5])
```
