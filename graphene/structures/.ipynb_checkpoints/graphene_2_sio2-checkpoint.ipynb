```python
from ase.io import read,write
import numpy as np
from ase.neighborlist import natural_cutoffs,NeighborList
import matplotlib.pyplot as plt
from ase import Atoms
from deepmd.calculator import DP
from ase.constraints import ExpCellFilter
from ase.optimize import LBFGS
from ase.calculators.lammpsrun import LAMMPS
import os
os.environ['ASE_LAMMPSRUN_COMMAND'] = 'lammps_mpi'
```

```python
pot = DP('../../coarse-grained-sio2/2D-silica-ML/Silica-dataset-generation/test/var0.1/frozen_9.pb')
```

```python
atom = read('9_9.vasp')
```

```python
cutoff = natural_cutoffs(atom)
nl  = NeighborList(cutoff, self_interaction=False, bothways=False,skin=0.1)
nl.update(atom)
```

```python
si_up = atom.get_positions()
si_down = atom.get_positions()
si_down[:,2] -= 3.3
O_medimu = atom.get_positions()
O_medimu[:,2] -= 1.65
O_up = []
for i in range(len(atom)):
    for j in nl.nl.neighbors[i]:
        O_up.append(atom.get_distance(i,j,mic=True,vector=True)/2+atom.get_positions()[i])
O_up = np.array(O_up)
O_down = O_up.copy()
O_down[:,2] -= 3.3 +0.5
O_up[:,2] += 0.5
new_pos = np.concatenate([si_up,si_down,O_medimu,O_up,O_down])

new_pos[:,:2] *= 2.28
cell = atom.get_cell()

cell[0,0] *= 2.28
cell[1,1] *= 2.28

new_atoms = Atoms(f'Si{si_up.shape[0]*2}O{len(O_medimu)+len(O_up)+len(O_down)}',positions=new_pos,cell=cell,pbc=True)
```

```python
si_up.shape,si_down.shape,O_medimu.shape,O_up.shape,O_down.shape
```

```python
(O_up.shape[0]+O_down.shape[0]+O_medimu.shape[0])/ (si_up.shape[0]*2)
```

```python
new_atoms.set_calculator(pot)
ecf = ExpCellFilter(new_atoms,[True,True,False,False,False,False])
dyn = LBFGS(new_atoms)
dyn.run(fmax=1e-2)
```

```python
write('sio2/99.xyz',new_atoms)
```
