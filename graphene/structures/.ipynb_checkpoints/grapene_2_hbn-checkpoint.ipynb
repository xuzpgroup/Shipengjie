```python
from ase.io import read,write
import numpy as np
from ase.neighborlist import natural_cutoffs,NeighborList
import matplotlib.pyplot as plt
```

```python
atom = read('9_9.vasp')
```

```python
cutoff = natural_cutoffs(atom)
nl  = NeighborList(cutoff, self_interaction=False, bothways=True)
nl.update(atom)
```

```python
Replacement  = set([0])
atomic_number = atom.get_atomic_numbers()
atomic_number[0] = 5
while len(Replacement)<len(atom):
    new_replace = set()
    for i in Replacement:
        neibor = nl.nl.neighbors[i]
        if len(neibor)==3:
            if atomic_number[i] == 5:
                for j in neibor:
                    atomic_number[j] =7
                    new_replace.add(j)
            if atomic_number[i] == 7:
                for j in neibor:
                    atomic_number[j] =5
                    new_replace.add(j)
        else:
            print(i)
    Replacement = Replacement.union(new_replace)
```

```python
for i in range(len(atom)):
    neibor = nl.nl.neighbors[i]
    if (atomic_number[i] == 5) and (atomic_number[neibor]==7).all():
        continue
    if (atomic_number[i] == 7) and (atomic_number[neibor]==5).all():
        continue
    print(i)
```

```python
atom.set_atomic_numbers(atomic_number)
```

```python
write('hbn/99.xyz',atom)
```

```python
from ase.constraints import ExpCellFilter,StrainFilter
from ase.optimize import BFGS,LBFGS
from ase.optimize.sciopt import SciPyFminCG
from deepmd.calculator import DP
```

```python
pot = DP('../../HBN/frozen_model_compressed.pb')
```

```python
atom = read('hbn/99.xyz')
atom.set_calculator(pot)
cell  = atom.get_cell()
cell[0,0] *= 1.02
cell[1,1] *= 1.02
atom.set_cell(cell,scale_atoms=True)
```

```python
ecf = ExpCellFilter(atom,[True,True,False,False,False,False])
```

```python
dyn = LBFGS(ecf)
```

```python
dyn.run(fmax=1e-2)
```
