```python
from ase.io import read, write
from ase.build import make_supercell, sort
from ase.neighborlist import natural_cutoffs, NeighborList
import numpy as np
import matplotlib.pyplot as plt
```

```python
atom = make_supercell(read('9_7.vasp'),[[59,0,0],[0,9,0],[0,0,1]])
```

```python
atom
```

```python
write('97_2000_500.xyz',atom)
```

```python
cell = atom.get_cell()
pos = atom.get_positions()
```

```python
cell[0,0] += 100
pos[:,0] += 50
```

```python
atom.set_cell(cell)
atom.set_positions(pos)
```

```python
cutoff = natural_cutoffs(atom)
nl = NeighborList(cutoff, self_interaction=False, bothways=True,skin=0)
nl.update(atom)
```

```python
I = []
for i in range(len(atom)):
    if nl.nl.neighbors[i].shape[0] <=1:
        I.append(i)
```

```python
len(I)
```

```python
del atom[I]
```

```python
write('97_2000_500_edge.xyz',atom)
```

```python
del atom[I]
```
